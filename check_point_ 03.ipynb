# Missing information and replace it with the appropriate values.

```python

df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
df['Cabin'].value_counts()
```

```python
df['Cabin'].fillna("N/A",inplace=True)
```

```python
df['Ticket'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder   #on accorde un chiffre pour chaque elt un code 
encode = LabelEncoder()
encode.fit(df['Ticket'])
```

```python
df['Ticket']=encode.transform(df['Ticket'])
```

```python
df['Sex']
```

```python
encoder={
    "male":0,
    "female":1
}
df["Sex"]=df["Sex"].apply(lambda x : encoder[x])
```

```python
df['Embarked'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder 
ohe = OneHotEncoder()
```

```python
enc_df = pd.DataFrame(ohe.fit_transform(df[['Embarked']]).toarray(),columns=ohe.get_feature_names())
enc_df
```

```python
df.isnull().sum()
```

```python
df['Embarked'].fillna('S',inplace=True)
enc_df
```

```python
df= df.join(enc_df)
df
```

# Studying the distribution of the most important features

```python
title_dictionary = {
                    "Capt"        :  "Officer",
                    "Col"         :  "Officer",
                    "Major"       :  "Officer",
                    "Dr"          :  "Officer",
                    "Rev"         :  "Officer",
                    "Jonkheer"    :  "Royalty",
                    "Don"         :  "Royalty",
                    "Sir"         :  "Royalty",
                    "Lady"        :  "Royalty",
                    "the Countess":  "Royalty",
                    "Dona"        :  "Royalty",
                    "Mme"         :  "Miss"   ,
                    "Mlle"        :  "Miss"   ,
                    "Miss"        :  "Miss"   ,
                    "Ms"          :  "Mrs"    ,
                    "Mr"          :  "Mrs"    ,
                    "Mrs"         :  "Mrs"    ,
                    "Master"      :  "Master"
                    }
```

```python
def recup_titre(x):
  part_string = x.split('.')[0]
  title = part_string.split(',')[1]
  title = title.strip()
  grade = title_dictionary[title]
  return grade
df['Title'] = df['Name'].apply(recup_titre)
```

```python
df.drop(axis=1,columns='Name',inplace=True)
```

```python
df['familySize']= df['SibSp']+df['Parch']+1
df['familySize']
```

```python
df.drop(axis=1,columns=['SibSp' ,'Parch'],inplace=True)
df.drop(axis=1,columns='Ticket',inplace=True)
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )

```
